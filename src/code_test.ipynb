```python
import os
import re
import pprint
from betacode_parser import *

content = {}

with open("../ccat/parallel/05.Deut.par") as file:
    word = 1
    for line in file:
        line = line.strip()
        if re.search(r'[1-9A-Z].+\d+:\d+', line): #Il faudra rcuprer qq par le book_name
            book, chap_num, verse_num = re.split(r'[\s:]', line)
            content.setdefault(book, {}).setdefault(chap_num, {}).setdefault(verse_num, {})
            word=1
        elif line:
            print(f"{chap_num}:{verse_num}:{word}:'{line}'")
            hb, gr = re.split(r'\t', line)
            hb = beta_to_hebrew(hb, True)
            hb = hb.replace("/", "")
            gr = beta_to_greek(gr)
            content[book][chap_num][verse_num][word] = {'hb': hb, 'gr': gr}
            word+=1
pprint.pprint(content) 
```
